```python
# generate random xy in circle
circle = Circle(xy=(0, 0), radius=2000)
circle_polygon = shapely.Polygon(circle.get_verts())
iot_xy = pointpats.random.poisson(circle_polygon, size=10)
iot_xyz = np.hstack((iot_xy, np.zeros(shape=(10, 1))))
iot_xyz.astype(int)
```

```python
fig, ax = plt.subplots()
ax.scatter(iot_xyz[:, 0], iot_xyz[:, 1])
for i in range(10):
    ax.annotate(str(i), (iot_xyz[i, 0], iot_xyz[i, 1]))
plt.title("Random IoT Location")
ax.set_xlim(-2000, 2000)
ax.set_ylim(-2000, 2000)
ax.set_aspect('equal')

plt.savefig('IoT_Genertated.pdf')
```
